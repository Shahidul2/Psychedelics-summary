```python
def clean_numeric(df, col, missing_val="Not Given"):
    """Converts to numeric after removing missing placeholders."""
    return pd.to_numeric(df[df[col] != missing_val][col], errors='coerce')

def compute_summary(df, name):
    # Clean numerical columns
    age = clean_numeric(df, 'age')
    weight = df['weight'].dropna()
    word_count = df['word_count'].dropna()

    # Gender distribution
    gender_counts = df['gender'].value_counts(normalize=True) * 100
    male_pct = gender_counts.get("Male", 0)
    female_pct = gender_counts.get("Female", 0)
    unspecified_pct = gender_counts.get("Not Specified", 0)

    return {
        "Substance": name,
        "Avg Age": age.mean(),
        "Median Age": age.median(),
        "Avg Weight (kg)": weight.mean(),
        "Median Weight (kg)": weight.median(),
        "Avg Word Count": word_count.mean(),
        "Median Word Count": word_count.median(),
        "% Male": male_pct,
        "% Female": female_pct,
        "% Not Specified": unspecified_pct
    }

# Compute summaries
summary_list = [
    compute_summary(df_mushroom, "Psilocybin"),
    compute_summary(df_lsd, "LSD"),
    compute_summary(df_dmt, "DMT")
]

# Create summary table
summary_df = pd.DataFrame(summary_list)

# Round for readability
summary_df = summary_df.round(2)

# Display
import pandas as pd
import IPython.display as disp
disp.display(summary_df)
```

```python
# 1. Merge all data
df_all = pd.concat([df_mushroom, df_lsd, df_dmt], ignore_index=True)

# 2. Helper to clean numeric column (same as before)
def clean_numeric(df, col, missing_val="Not Given"):
    return pd.to_numeric(df[df[col] != missing_val][col], errors='coerce')

# 3. Summary function for combined data
def compute_overall_summary(df):
    age = clean_numeric(df, 'age')
    weight = df['weight'].dropna()
    word_count = df['word_count'].dropna()

    gender_counts = df['gender'].value_counts(normalize=True) * 100
    male_pct = gender_counts.get("Male", 0)
    female_pct = gender_counts.get("Female", 0)
    unspecified_pct = gender_counts.get("Not Specified", 0)

    return pd.DataFrame([{
        "Substance": "All Substances Combined",
        "Avg Age": age.mean(),
        "Median Age": age.median(),
        "Avg Weight (kg)": weight.mean(),
        "Median Weight (kg)": weight.median(),
        "Avg Word Count": word_count.mean(),
        "Median Word Count": word_count.median(),
        "% Male": male_pct,
        "% Female": female_pct,
        "% Not Specified": unspecified_pct
    }]).round(2)

# 4. Run it
summary_all_df = compute_overall_summary(df_all)

# 5. Display the summary
import IPython.display as disp
disp.display(summary_all_df)
```

```python
from scipy.stats import skew, kurtosis

def enhanced_summary_stats(series):
    series = pd.to_numeric(series, errors='coerce').dropna()
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    return {
        "Count": len(series),
        "Mean": series.mean(),
        "Median": series.median(),
        "Std Dev": series.std(),
        "Min": series.min(),
        "25th %ile": q1,
        "75th %ile": q3,
        "Max": series.max(),
        "IQR": q3 - q1,
        "Skewness": skew(series),
        "Kurtosis": kurtosis(series)
    }

# Compute for each column
age_stats = enhanced_summary_stats(df_all[df_all['age'] != 'Not Given']['age'])
weight_stats = enhanced_summary_stats(df_all['weight'])
word_stats = enhanced_summary_stats(df_all['word_count'])

# Gender breakdown
gender_counts = df_all['gender'].value_counts(normalize=True) * 100

# Combine everything into one row DataFrame
summary_full = pd.DataFrame({
    "Metric": list(age_stats.keys()),
    "Age": list(age_stats.values()),
    "Weight (kg)": list(weight_stats.values()),
    "Word Count": list(word_stats.values())
})

# Add gender stats
gender_row = pd.DataFrame({
    "Metric": ["% Male", "% Female", "% Not Specified"],
    "Age": [None]*3,
    "Weight (kg)": [None]*3,
    "Word Count": [None]*3
})
gender_row.loc[0, 'Metric'] = "% Male"
gender_row.loc[0, 'Word Count'] = gender_counts.get("Male", 0)
gender_row.loc[1, 'Metric'] = "% Female"
gender_row.loc[1, 'Word Count'] = gender_counts.get("Female", 0)
gender_row.loc[2, 'Metric'] = "% Not Specified"
gender_row.loc[2, 'Word Count'] = gender_counts.get("Not Specified", 0)

# Final summary
summary_final = pd.concat([summary_full, gender_row], ignore_index=True)
summary_final = summary_final.round(2)

# Display
import IPython.display as disp
disp.display(summary_final)
```
